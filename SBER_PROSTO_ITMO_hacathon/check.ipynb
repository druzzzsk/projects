```python
import pandas as pd
```

```python
df = pd.read_csv('./generated.csv')
df.head()
```

```python
df.shape
```

```python
def replace_third_number(tup_str):
    numbers = list(map(float, tup_str.strip('()').split(',')))
    numbers[2] = 0
    new_tup_str = f"({', '.join(map(str, numbers))})"
    return new_tup_str

df['Fitness Values'] = df['Fitness Values'].apply(replace_third_number)
```

```python
df.head()
```

```python
def extract_first_three(tup_str):
    numbers = tuple(map(float, tup_str.strip('()').split(',')))
    return numbers[:3]

df[['Value1', 'Value2', 'Value3']] = df['Fitness Values'].apply(extract_first_three).apply(pd.Series)
```
